### Splitting the dataset into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=10) #Try to use 'stratify = True' for better splitting 
```

### Feature Scailing

* The algorithms having gradient discent(LinearRegression),Eucledian distance(knn) requires feature scailing. 
The data points should be in the graph limitation range so we scale the features magnitude to be fitted in the graph.

* Ensemble techniques(decision forest, random forest, xg boost) dont require feature scailing. 


*** Standardization vs Normalization ***

The main purpose of this scales are to rescale the values of features in the same unit so that in the obtained distribution every point or value is close to each other.
* standardization and normalization are scailing techniques which are applied on an algorithm to make features of data into our required scale(i.e units and magnitude). So, that features become easy to apply on the model.

* Units and magnitude 
(ex:25yrs --> yrs here is units and 25 is our magnitude)



* Standardization: it helps to scale down the features of data based on standard and normal distribution curve with mean=0 & standard deviation=1

  - range: bell curve
  - Xstand = (X-mean) / (sd)
  - StandardScaler() 
also called as Z-Score Normalization


* Normalization: it helps to scale down features of data between 0 to 1

   - range : 0 to 1
   - Xnorm = (X-Xmin) / (Xmax-Xmin)
   - MinMaxScaler() 

* Reference Link : https://www.analyticsvidhya.com/blog/2020/04/feature-scaling-machine-learning-normalization-standardization/

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
#r2 - performance metric for regression problem
```

```python
y_pred = lr.predict(x_test)
y_pred
```

### Performance Metric for Regression Problem(r2_score)

```python
from sklearn.metrics import r2_score
r2_scoreLR = r2_score(y_test, y_pred)
```

```python
r2_scoreLR
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split

max_features = ['auto', 'sqrt']
# Maximum number of levels in tree

max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]

# Method of selecting samples for training each tree
bootstrap = [True, False]

# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python

from sklearn.model_selection import RandomizedSearchCV
rfr1=RandomForestRegressor()

rf_random = RandomizedSearchCV(estimator = rfr1, 
                               param_distributions = random_grid, 
                               n_iter = 100, 
                               cv = 3, 
                               verbose=2, 
                               random_state=42, 
                               n_jobs = -1)


rf_random.fit(x_train, y_train)
```

```python
rf_random.best_params_
```

```python
best_random = rf_random.best_estimator_
```
